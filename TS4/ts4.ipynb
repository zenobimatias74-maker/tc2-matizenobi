## Transferencias de Bessel con método de Storch

### Introducción

Los filtros de **Bessel**, a diferencia de máxima planicidad o Chebyshev, se centran en la fase y el tiempo. La aproximación de Bessel logra una fase específica planteando al polinomio divisor del filtro pasa bajos de forma tal que el cociente de su parte imaginaria y real provenga de la tangente de la fase.

Con un **retardo** de 1 segundo, $D(\omega=0)=1\text s$:

$$
D(\omega)=-\frac{d\theta(\omega)}{d\omega}=1\Longrightarrow\theta(\omega)=-\omega
\newline
\tan\big(\theta(\omega)\big)=\tan(-\omega)=-\tan(\omega)

$$

El **método de Storch** es un algoritmo para facilitar el cálculo de dicha tangente.

$$
\tan(\omega)=\frac{\omega}{1-\frac{\omega^2}{3-\frac{\omega^2}{5-\text{...}}}}
$$

---

### Orden 2

$$
T(j\omega)=\frac{a_0}{(j\omega)^2+a_1j\omega+a_0}=\frac{a_0}{(a_0-\omega^2)+j(a_1\omega)}\newline
\tan(\omega)=\frac{a_1\omega}{a_0-\omega^2}
$$

(Se cancelan los signos negativos ya que es la fase del denominador)

Para $N=2$, la expansión se corta en la segunda iteración.

$$
\tan(\omega)=\frac{\omega}{1-\frac{\omega^2}{3}}=\frac{3\omega}{3-\omega^2}
$$

$$
T(s)=T(j\omega)\Big|_{\omega=s/j}=\frac{3}{s^2+3s+3}
$$

### Orden 3

\newline
\tan(\omega)=\frac{a_1\omega-\omega^3}{a_0-a_2\omega^2}
$$

Para $N=3$, la expansión se corta en la tercera iteración:

$$

$$

$$

$$

## Retardo en $D(\omega=2.5)$

$$

\tan(\omega)=-\tan\big(\theta(\omega)\big)=\frac{15\omega-\omega^3}{15-6\omega^2}
\Longrightarrow\theta(\omega)=-\arctan\Big(\frac{15\omega-\omega^3}{15-6\omega^2}\Big)\newline
D(\omega)=-\frac{d\theta(\omega)}{d\omega}=\frac{d}{d\omega}\Big[\arctan\Big(\frac{15\omega-\omega^3}{15-6\omega^2}\Big)\Big]=\frac{6\omega^4+45\omega^2+225}{\omega^6+6\omega^4+45\omega^2+225}

### Sección de segundo orden

![imagen.png](imagen%201.png)

$$
\begin{cases}
\begin{align*}
\Big(sC_2+\frac{1}{R_1}+\frac{1}{R_2}\Big)V_A(s)-sC_2V_o(s)-\frac{1}{R_1}V_i(s)-\frac{1}{R_2}V_B(s)&=0\newline
\Big(sC_1+\frac{1}{R_2}\Big)V_B(s)-\frac{1}{R_2}V_A(s)&=0\newline
V_B(s)&=V_o(s)
\end{align*}
\end{cases}
$$

```python
import sympy as sp
from pytc2.general import a_equal_b_latex_s, print_latex, s

Vi, Vo, Va, Vb = sp.symbols('V_i V_o V_A V_B')
Y1, Y2, G1, G2 = sp.symbols('Y_{C_1} Y_{C_2} G_1 G_2')
C1, C2, R1, R2 = sp.symbols('C_1 C_2 R_1 R_2')

nodoA = sp.Equality((Y2+G1+G2)*Va-Y2*Vo-G1*Vi-G2*Vb, 0)
nodoB = sp.Equality((Y1+G2)*Vb-G2*Va, 0)
retro = sp.Equality(Vb, Vo)

solution = sp.solve((nodoA, nodoB, retro), [Vo, Va, Vb])

# Transferencia en función de admitancias:
T_y = sp.simplify(solution[Vo]/Vi)
#print_latex(a_equal_b_latex_s('T_Y(s) = \\frac{V_o(s)}{V_i(s)}', T_y))

# Transferencia en función de componentes y frecuencia compleja (s)
T = sp.simplify(T_y.subs({Y1:s*C1, Y2:s*C2, G1:1/R1, G2:1/R2}))
print_latex(a_equal_b_latex_s('T(s) = \\frac{V_o(s)}{V_i(s)}', T))
```
